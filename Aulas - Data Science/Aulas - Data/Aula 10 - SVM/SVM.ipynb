SVM

* Modelo de classificação e regressão supervisionados

* Classificador binário não probabilístico

* É representado pela separação dos dados a partir de um hiperplano

Maximum Street Approach

Método utilizado para encontrar o hiperplano que maximize as margens entre as classes.

As margens devem possuir comprimentos iguais em ambos os lados do hiperplano.

Classificação SVM

Eq. geral do plano: $ax + by + cy + d = 0$

Projeção p de u em w:



$$ p = \alpha \cdot w$$
$$p = \frac{w \cdot u}{||w||²}\cdot w$$
$$||p|| = \frac{|u \cdot w|}{||w||}$$



Tamanho da margem: 
$$\frac{2}{||W||}$$

Objetivo: maximizar a margem, portanto a expressão acima.

Para isto, minimizar $||W||$ e, da mesma forma, minimizar $||W^2||$

$$min(\frac{1}{2}||W²||)$$

restrições: 
$$y_i \cdot (W \cdot x + b) \geq 1$$
$$p_x ||w|| \geq 1$$
$$p_x ||w|| < -1$$

Hard Margin vs Soft Margin

Não é vantajoso, por risco de overfitting, que o hiperplano separe os dados linearmente.



Portanto, um erro deverá ser associado e uma Soft margin é melhor utilizável.

SVM é o melhor algoritmo se for possível uma separação linear dos dados.

Soft Margin: 

$$y_i (Wx + b) \geq 1 - \epsilon_i$$

$$min(||W||) \longrightarrow min(||W||) + C\sum_{i= 1}^{m}{\epsilon_i}$$

Influência do hiperparâmetro C

* Quanto menor C, menor o custo de classificar errado. Constrói margem grande

Kernels

* Corresponde a aplicação de uma função/transformação que muda o espaço dos nossos dados

* No novo espaço, será possível separar os dados linearmente

Exemplo - Kernel Polinomial de grau 2

$$K(u,v) = \Phi(u) \cdot \Phi(v)$$
$$\Phi(x) = \Phi((x_1, x_2)) = (x_1², \sqrt{2} \cdot x_1^2, x_2²)$$

 Kernel | Equation
--------|----------
 Linear | $K(x,y) = x \cdot y$
 Sigmoid| $K(x,y) = tanh(axy + b)$
 Polynomial| $K(x,y) = (1 + xy)^d$
 KMOD| $K(x,y) = a[exp(\frac{\gamma}{\|\|x-y\|\|² + \sigma²})]$
 RBF| $K(x,y) = exp(-a\|\|x-y\|\|²)$
 Exponential RBF| $K(x,y) = exp(-a\|\|x-y\|\|)$

SVM para Regressões

Construção de hiperplano que abrange os dados, ao invés de separá-los

Vantagens do modelo SVM

* Predição eficiente em tempo e memória

* Funciona bem com alta dimensionalidade

* Reduz overfitting

* Classificação não linear com kernel

* Se encontrar o kernel ótimo, será o melhor modelo possível

Desvantagens do modelo SVM

* Complexidade de treino alta

* Não fornece probabilidades (1)

* Não é possível realizar classificação multiclass (2)

(1) - Modelos alternativos utilizam distância ao hiperplano para estimar probabilidades

(2) - One vs All - SVM para classificação de cada classe
