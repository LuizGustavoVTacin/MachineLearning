Problemas em mtodos de ML:

Classificador Enviesado: No possuem capacidade para modelar a funo verdadeira. 

Uma rvore de deciso enviesada no consegue separar os dados de forma correta



Classificadores enviesados tem a vantagem de no precisar de muitos dados para convergir para uma soluo, j que no sofrem com overfitting

Apesar disso, eles sofrem do underfitting, j que no aprendem funes complexas

Classificadores com alta varincia:

Este tipo de classificador consegue aprender funes altamente complexas, chegando at ao erro 0.

Vantagens: Pode encontrar uma soluo com acurcia muito boa para o problema.

Desvantagens: O modelo incorpora padres aleatrios do dataset que no representa a realidade

rvore de Deciso com varincia



A soluo para o enviesamento e as altas varincias  o balanceamento ou utilizar vrios classificadores com enviesamento ou alta varincia para um nico modelo.

Para isto, seria necessrio mais de um dataset para o treinamento de mais modelos. Desta forma, poderia-se usar um dataset para cada modelo e, ao final, calcula-se a mdia de todos os modelos.

Boostrapping

Os conjuntos derivados sero obtidos a partir da amostragem de elementos aleatrios do original

No  possvel fazer amostragem de um dataset sem reposio pois gerar o mesmo dataset.

Random Forests

No uso de vrias rvores de deciso treinadas com bootstrap,  necessrio que os erros sejam compensados. Para isto,  essencial que exista diversidade entre os modelos.

Todas as rvores que tiveram acesso as mesmas features, tero as mesmas caractersticas. Por exemplo, se a feature por preditiva, ento todas as rvores iro us-la, diminuindo a diversidade

Random forest  uma variao do Bagging que aumentar essa diversidade. Cada rvore no ter acesso a todas as features. Ao invs disso, seleciona-se k features para cada uma que sero as nicas visveis para o modelo.

O valor de k  um hiperparmetro, mas $\sqrt{Num_{features}}$  um bom valor inicial

O clculo do erro  feito com a mtrica desejada sobre pontos do dataset que as rvores no tiveram acesso. Ao final, basta fazer a mdia dos erros em todos os pontos.
