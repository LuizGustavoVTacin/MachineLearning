Problemas em métodos de ML:

Classificador Enviesado: Não possuem capacidade para modelar a função verdadeira. 

Uma árvore de decisão enviesada não consegue separar os dados de forma correta



Classificadores enviesados tem a vantagem de não precisar de muitos dados para convergir para uma solução, já que não sofrem com overfitting

Apesar disso, eles sofrem do underfitting, já que não aprendem funções complexas

Classificadores com alta variância:

Este tipo de classificador consegue aprender funções altamente complexas, chegando até ao erro 0.

Vantagens: Pode encontrar uma solução com acurácia muito boa para o problema.

Desvantagens: O modelo incorpora padrões aleatórios do dataset que não representa a realidade

Árvore de Decisão com variância



A solução para o enviesamento e as altas variâncias é o balanceamento ou utilizar vários classificadores com enviesamento ou alta variância para um único modelo.

Para isto, seria necessário mais de um dataset para o treinamento de mais modelos. Desta forma, poderia-se usar um dataset para cada modelo e, ao final, calcula-se a média de todos os modelos.

Boostrapping

É um método que permite, a partir de um único dataset, obter vários outros conjuntos derivados do mesmo tamanho do original e com mesma distribuição de dados.

Os conjuntos derivados serão obtidos a partir da amostragem de elementos aleatórios do original

Amostragem com reposição:

Não é possível fazer amostragem de um dataset sem reposição pois gerará o mesmo dataset.

Em média, 63,2% dis itens serão derivados do original

Algoritmo que implementa o Bootstrap

Random Forests

No uso de várias árvores de decisão treinadas com bootstrap, é necessário que os erros sejam compensados. Para isto, é essencial que exista diversidade entre os modelos.

Todas as árvores que tiveram acesso as mesmas features, terão as mesmas características. Por exemplo, se a feature por preditiva, então todas as árvores irão usá-la, diminuindo a diversidade

Random forest é uma variação do Bagging que aumentará essa diversidade. Cada árvore não terá acesso a todas as features. Ao invés disso, seleciona-se k features para cada uma que serão as únicas visíveis para o modelo.

O valor de k é um hiperparâmetro, mas $\sqrt{Num_{features}}$ é um bom valor inicial

Erro Out-of-Bag

Não é necessário separar parte do dataset para validação, já que nem todos os modelos viram os mesmos pontos.

O cálculo do erro é feito com a métrica desejada sobre pontos do dataset que as árvores não tiveram acesso. Ao final, basta fazer a média dos erros em todos os pontos.



Apesar de não ser necessário o conjunto de validação, é necessário o conjunto de teste

Estimação de Incerteza

A incerteza é calculada para classificar uma classe com base em uma probabilidade de pertencimento

Extra Trees

Variação do Random Forests que aumenta a variabilidade dos modelos na ensemble. Para cada feature, ao invés de testarmos cada posição do split para encontrar a melhor divisão dos dados, utiliza-se uma aleatória.
