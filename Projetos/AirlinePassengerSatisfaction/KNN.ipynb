```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## **Importing Datasets**

```python
df_test = pd.read_csv('Dataset/test.csv', index_col=0)
df_train = pd.read_csv('Dataset/train.csv', index_col=0)

df_test = pd.DataFrame(df_test)
df_train = pd.DataFrame(df_train)

df_test.tail()
```

```python
df_train.tail()
```

## **Removing undesired columns**

**id** is not a subject that will influence the satisfaction of the passenger, so it will be removed.

```python
df_test = df_test.drop(['id'], axis=1)
df_train = df_train.drop(['id'], axis=1)
```

## **Checking data**

```python
df_train.describe()
```

```python
df_train.isna().sum()
```

NaN's in Arrival Delay in Minutes are intepreted as no occurrence of delay, so it will be replaced by 0.

```python
df_train['Arrival Delay in Minutes'].replace(np.nan, 0, inplace=True)
df_test['Arrival Delay in Minutes'].replace(np.nan, 0, inplace=True)
```

## **Replacing Strings by Numbers**

```python
df_train.head()
```

```python
columns_string = ['Gender','Customer Type','Type of Travel','Class']
def Factorize(df):
    for column in columns_string:
        labels, uniques = pd.factorize(df[column])
        mapping = {label: string for label, string in zip(uniques, labels)}
        df[column] = df[column].map(mapping)
    return df

df_train = Factorize(df_train)
df_test = Factorize(df_test)
df_train.head()
```

## **Spliting Training Data and Desired Outputs**

```python
df_test_Satisfaction = df_test['satisfaction']
df_test = df_test.drop('satisfaction', axis=1)
df_train_Satisfaction = df_train['satisfaction']
df_train = df_train.drop('satisfaction', axis=1)
```

## **Ajusting Data**

```python
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler()
df_test = pd.DataFrame(Scaler.fit_transform(df_test))
df_train = pd.DataFrame(Scaler.fit_transform(df_train))

df_test.head()
```

## **Saving DataFrames in CSV files for future use**

```python
df_test.to_csv('Dataset/test_ajusted.csv')
df_train.to_csv('Dataset/train_ajusted.csv')
df_test_Satisfaction.to_csv('Dataset/test_Satisfaction.csv')
df_train_Satisfaction.to_csv('Dataset/train_Satisfaction.csv')
```

## **Creating Validation Data**

```python
from sklearn.model_selection import train_test_split

X = df_train
Y = df_train_Satisfaction
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

## **Training Model**

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=5)
train = KNN.fit(X_train, Y_train)
Result = KNN.predict(X_test)
```

## **Testing with Validation Data**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import seaborn as sns

print(accuracy_score(Result, Y_test))
print(classification_report(Result, Y_test))

cm = confusion_matrix(Result, Y_test)
names = np.unique(Result)
sns.heatmap(cm, square= True, annot= True, fmt= 'd', cbar= False, xticklabels= names, yticklabels= names)
plt.xlabel('Truth')
plt.ylabel('Predicted')
```
