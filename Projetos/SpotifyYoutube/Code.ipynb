## **Handling Data**

### **Counting Null Values**

```python
df.isna().sum()
```

NaN on Views, Likes, Stream and Comments are interpreted as no occurrences of these events on the video so will be replaced with 0

```python
df['Views'] = df['Views'].replace({np.nan: 0})
df['Likes'] = df['Likes'].replace({np.nan: 0})
df['Stream'] = df['Stream'].replace({np.nan: 0})
df['Comments'] = df['Comments'].replace({np.nan: 0})
df.isna().sum()
```

Title, Channel, Licensed and official_video are attibutes that belongs to youtube, meaning that the correspondent rows have musics that are only on spotify and not on youtube.

```python
## Checking if all the NaN values are in the same position. If so, all the musics with NaN = 470 only belongs to Spotify.
NaN_positions_Title = np.where(df['Title'].isna())[0]
NaN_positions_Channel = np.where(df['Channel'].isna())[0]
NaN_positions_Licensed = np.where(df['Licensed'].isna())[0]
NaN_positions_official_video = np.where(df['official_video'].isna())[0]

vecTrue = NaN_positions_Title == NaN_positions_Channel
vecTrue1 = NaN_positions_Title == NaN_positions_Licensed
vecTrue2 = NaN_positions_Title == NaN_positions_official_video

if vecTrue.all() == True and vecTrue1.all() == True and vecTrue2.all() == True:
    print('True')
```

Danceability, Energy, Key, Loudness, Mode, Speechiness, Acousticness, Instrumentalness, Liveness, Valence, Tempo, Duration_ms have the same amount of NaN and will be considered as the same rows.

## **Top 10 Artists on Youtube and Spotify**

```python
df_youtube = df.groupby('Artist')['Views'].sum()
df_spotify = df.groupby('Artist')['Stream'].sum()
df_youtube_sorted = df_youtube.sort_values(ascending= False)[0:10][::-1] ## [::-1] is to reverse the order, so the highest value will be on top
df_spotify_sorted = df_spotify.sort_values(ascending= False)[0:10][::-1]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize= (12,6))
ax1.set_title('Top 10 Artists on Youtube')
df_youtube_sorted.plot(kind= 'bar', ax= ax1)
ax2.set_title('Top 10 Artists on Spotify')
df_spotify_sorted.plot(kind= 'bar', ax=ax2)

ax1.set_xlabel('Artist')
ax1.set_ylabel('Views')
ax2.set_xlabel('Artist')
ax2.set_ylabel('Streams')
fig.tight_layout() #Automatic Adjustments for the subplots positions
plt.show()
```

## **Top 10 Songs on Youtube and Spotify**

```python
df_Trackyoutube = df.nlargest(12, 'Views').drop_duplicates(subset= ['Track'])[::-1] #[::-1] inverts the series
df_Trackspotify = df.nlargest(13, 'Stream').drop_duplicates(subset= ['Track'])[::-1]

## nlargest instead of the previous groupy method is better because of the duplicates on spotify and youtube sums
## drop_duplicates is used to remove the duplicates on the Track column. Duplicates are ocurring on musics with 2 or more artists.

fig, (ax1, ax2) = plt.subplots(2, 1, figsize= (12,6))
ax1.set_title('Top 10 Artists on Youtube')
df_Trackyoutube.plot(kind= 'barh', ax= ax1, x= 'Track', y= 'Views')
ax2.set_title('Top 10 Artists on Spotify')
df_Trackspotify.plot(kind= 'barh', ax=ax2, x= 'Track', y='Stream')
ax1.set_title('Top 10 Tracks on youtube')
ax2.set_title('Top 10 Tracks on Spotify')
fig.tight_layout() #Automatic Adjustments for the subplots positions
plt.show()
```

## **Top 100 Tracks Views and Streams percentage**

```python
df_Trackspotify100 = df.nlargest(119, 'Stream').drop_duplicates(subset= ['Track'])
labels = ['Top 100 most \n Streamed', 'All']
colors = ['red', 'blue']
explode = [0.3, 0]
values = [df_Trackspotify100['Stream'].sum(), df['Stream'].sum()]
plt.pie(values, labels= labels, colors= colors, autopct= '%1.1f%%', explode=explode)
plt.show()
```

```python
df_Trackspotify100 = df.nlargest(121, 'Views').drop_duplicates(subset= ['Track'])
print(df_Trackspotify100.shape)
labels = ['Top 100 most \n Viewed', 'All']
colors = ['green', 'brown']
explode = [0.3, 0]
values = [df_Trackspotify100['Stream'].sum(), df['Stream'].sum()]
plt.pie(values, labels= labels, colors= colors, autopct= '%1.1f%%', explode=explode, shadow= True)
plt.show()
```

## **Top 10 Longest Tracks**
