```python
df = df.drop(['Url_spotify', 'Url_youtube', 'Uri', 'Title', 'Description'], axis= 1)
df.tail()
```

#### Checking Null Values

```python
df.isna().sum()
```

NaN on Views, Likes, Stream and Comments are interpreted as no occurrences on these events on the video so will be replaced with 0

```python
change = {np.nan: 0}
df['Views'] = df['Views'].replace(change)
df['Likes'] = df['Likes'].replace(change)
df['Stream'] = df['Stream'].replace(change)
df['Comments'] = df['Comments'].replace(change)
df.isna().sum()
```
