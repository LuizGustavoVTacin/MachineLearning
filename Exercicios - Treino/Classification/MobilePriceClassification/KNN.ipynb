```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Import Train

```python
df = pd.read_csv('train.csv')
df = pd.DataFrame(df)
df.tail()
```

## Import Test

```python
df_test = pd.read_csv('test.csv')
df_test = pd.DataFrame(df_test)
df_test = df_test.drop('id', axis= 1)
df_test.tail()
```

```python
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler()
df_resul = df['price_range']
df = Scaler.fit_transform(df.drop('price_range', axis= 1))
df = pd.DataFrame(df)
df.head()
```

## Spliting Train Dataset between Train and Validation

```python
from sklearn.model_selection import train_test_split

X = df
Y = df_resul

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier()
Train = KNN.fit(X_train, Y_train)
Result = Train.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print(accuracy_score(Result, Y_test))
print(classification_report(Result, Y_test))

cm = confusion_matrix(Result, Y_test)
names = np.unique(Result)
sns.heatmap(cm, square= True, annot= True, fmt= 'd', cbar= False, xticklabels= names, yticklabels= names)
plt.xlabel('Truth')
plt.ylabel('Predicted')
```
