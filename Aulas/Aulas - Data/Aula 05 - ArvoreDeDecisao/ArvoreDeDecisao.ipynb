Árvores de Decisão:

Método de aprendizado de máquina supervisionado utilizado em tarefas de classificação e regressão.

Cada ramificação na árvore descreve caminhos de decisões.

Flexibilidade do modelo:

* Tipos de dados:

Um nó de decisão dentro da árvore pode utilizar uma feature categórica ou numérica

* Sub-árvores:

As sub-árvores à esquerda e à direita de um nó de decisão não necessariamente seguem a mesma ordem em relação à features

Construindo uma árvore de decisão

O início da construção envolve a divisão espacial das classes em linhas.

A medida da impureza dirá quão bem as divisões foram feitas. Quanto menor a impureza, melhor o algoritmo

Cálculos de impureza:

* Features categóricos: Impureza de Gini e Entropia

* Features numéricas: Redução da variância 

Impureza de Gini:

Corresponde a verossimilhança de uma classificação.

$$G(s) = \sum_{k= 1}^{K}p_k (1 - p_k)$$

Entropia:

Mede a desordem de um sistema. 

Quanto maior a entropia, maior a quantidade de diferentes classes de dados. 

Quando há maior homogeneidade entre as classes, menor a entropia

$$H(p_k) = - \sum^K_{k= 1}{p_k log(p_k)}$$

Variance Reduction:

Quanto mais próximos os valores numéricos dentro de uma região descrita pelo corte, menor o valor computado por variance reduction

$$L(s) = \frac{1}{|S|} \sum_i^S{(y_i - \mu_S)²}$$

Cálculo da pureza total:

$$Impureza = \frac{|S_L|}{|S|} M(S_L) + \frac{|S_R|}{|S|}M(S_R)$$

M = Uma das métricas anteriores

SR = Conjunto de dados do lado direito do corte

SL = conjunto de dados do lado esquerdo do corte

Custo computacional do treino:

$$O(ND \cdot KN) = O(KDN²)$$

K = Número de classes

N = Número de dados

D = Número de features

Árvore Ideal:

Uma árvore muito profunda tende ao overfit

Uma árvore muito curta tende ao underfit

Vantagens de Árvores de decisão:

* Fácil interpretação: descemos uma árvore e escolhemos o caminho de acordo com as features

* Funcionam com qualquer tipo de dado

* Não dependem de normalização ou scaling dos dados

* Excessivamente rápidas e otimizadas

* É utilizado de base para outros modelos melhores

Desvantagens de Árvores de decisão:

* É complicado encontrar exatamente a árvore com profundidade ótima

* Uma faixa muito restrita de profundidade minimiza os erros em relação ao conjunto de teste e ao conjunto de treino
